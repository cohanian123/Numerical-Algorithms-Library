#**Root-Finding Algorithms**

```python
# REGULA FALSI

def regula_falsi(f, a, b, rho, epsilon=1e-6, max_iter=1000):
    # Parameters:
        # f (callable function): Function we are trying to find the root of
        # a (float): Initial lower bound of the interval
        # b (float): Initial upper bound of the interval
        # rho (float): True solution
        # epsilon (float): Convergence criterion
        # max_iter (int): Maximum number of iterations
    # Returns:
        # x_next (float): Computed root
        # iterations (list): List of tuples with (iteration, x, delta_x, relative_error)

    if f(a) * f(b) >= 0:
        raise ValueError("f(a) and f(b) must have opposite signs.")

    q = (f(b) - f(a)) / (b - a)    # Initial q
    x = b    # Initial guess
    k = 0    # Initialize iteration counter
    iterations = []  # Store iteration details

    while k < max_iter:
        x_next = x - f(x) / q    # Compute next x
        delta_x = x_next - x  # Change in x
        error_next = x_next - rho
        error = x - rho
        relative_error = error_next / error if error != 0 else np.nan  # Avoid division by zero
        iterations.append((k + 1, x_next, delta_x, relative_error))  # Log the iteration

        if abs(error_next) < epsilon:
            return x_next, iterations  # Convergence check

        # Update the interval based on the sign of f(x_next)
        if f(x_next) * f(a) < 0:
            b = x_next
        elif f(x_next) * f(b) < 0:
            a = x_next
        else:
            return x_next, iterations  # Convergence check

        q = (f(b) - f(a)) / (b - a)    # Update q
        x = x_next    # Update x
        k += 1

    raise ValueError("Regula Falsi method did not converge within maximum number of iterations")
```

```python
# SECANT METHOD

def secant(f, x0, x1, rho, epsilon=1e-6, max_iter=1000):
    # Parameters:
        # f (callable function): Function we are trying to find the root of
        # x0 (float): Initial guess x_{-1}
        # x1 (float): Initial guess x_{0}
        # rho (float): True solution
        # epsilon (float): Convergence criterion
        # max_iter (int): Maximum number of iterations
    # Returns:
        # x_next (float): Computed root
        # iterations (list): List of tuples with (iteration, x, delta_x, relative_error)

    k = 0  # Initialize iteration counter
    iterations = []  # Store iteration details

    while k < max_iter:
        q = (f(x1) - f(x0)) / (x1 - x0)    # Compute q_k
        x_next = x1 - f(x1) / q    # Compute next x
        delta_x = x_next - x1  # Change in x
        error_next = x_next - rho
        error = x1 - rho
        relative_error = error_next / error if error != 0 else np.nan  # Avoid division by zero
        iterations.append((k + 1, x_next, delta_x, relative_error))  # Log the iteration

        if abs(error_next) < epsilon:
            return x_next, iterations  # Convergence check

        x0 = x1    # Update x values
        x1 = x_next
        k += 1

    raise ValueError("Secant Method did not converge within the maximum number of iterations.")
```

```python
# NEWTON'S METHOD

def newton(f, df, x0, rho, epsilon=1e-6, max_iter=1000):
    # Parameters:
        # f (callable function): Function we are trying to find the root of
        # df (callable function): Derivative of f
        # x0 (float): Initial guess
        # rho (float): True solution
        # epsilon (float): Convergence criterion
        # max_iter (int): Maximum number of iterations
    # Returns:
        # x_next (float): Computed root
        # iterations (list): List of tuples with (iteration, x, delta_x, relative_error)

    k = 0  # Initialize iteration counter
    x = x0  # Initial guess
    iterations = []  # Store iteration details

    while k < max_iter:
        q = df(x)  # Compute q_k (the derivative at the current point)
        if q == 0:
            raise ValueError(f"Derivative is zero at iteration {k}. Method fails.")

        x_next = x - f(x) / q  # Compute next x
        delta_x = x_next - x  # Change in x
        error_next = x_next - rho
        error = x - rho
        relative_error = error_next / error if error != 0 else np.nan  # Avoid division by zero
        iterations.append((k + 1, x_next, delta_x, relative_error))  # Log the iteration

        if abs(error_next) < epsilon:
            return x_next, iterations  # Convergence check

        x = x_next  # Update x
        k += 1

    raise ValueError("Newton's Method did not converge within the maximum number of iterations.")
```

```python
# MODIFIED NEWTON'S METHOD

def mod_newton(f, df, x0, rho, m, epsilon=1e-6, max_iter=1000):
    # Parameters:
        # f (callable function): Function we are trying to find the root of
        # df (callable function): Derivative of f
        # x0 (float): Initial guess
        # rho (float): True solution
        # m (float): Positive integer value corresponding to the multiplicity of the root sought
        # epsilon (float): Convergence criterion
        # max_iter (int): Maximum number of iterations
    # Returns:
        # x_next (float): Computed root
        # iterations (list): List of tuples with (iteration, x, delta_x, relative_error)

    k = 0  # Initialize iteration counter
    x = x0  # Initial guess
    iterations = []  # Store iteration details

    while k < max_iter:
        q = df(x)  # Compute q_k (the derivative at the current point)
        if q == 0:
            raise ValueError(f"Derivative is zero at iteration {k}. Method fails.")

        x_next = x - m * (f(x) / q)  # Compute next x
        delta_x = x_next - x  # Change in x
        error_next = x_next - rho
        error = x - rho
        relative_error = error_next / error if error != 0 else np.nan  # Avoid division by zero
        iterations.append((k + 1, x_next, delta_x, relative_error))  # Log the iteration

        if abs(error_next) < epsilon:
            return x_next, iterations  # Convergence check

        x = x_next  # Update x
        k += 1

    raise ValueError(" Modified Newton's Method did not converge within the maximum number of iterations.")
```

```python
# STEFFENSEN'S METHOD

def steffensen(f, x0, rho, epsilon=1e-6, max_iter=1000):
    # Parameters:
        # f (callable function): Function we are trying to find the root of
        # x0 (float): Initial guess
        # rho (float): True solution
        # epsilon (float): Convergence criterion
        # max_iter (int): Maximum number of iterations
    # Returns:
        # x_next (float): Computed root
        # iterations (list): List of tuples with (iteration, x, delta_x, relative_error)

    k = 0  # Initialize iteration counter
    x = x0  # Initial guess
    iterations = []  # Store iteration details

    while k < max_iter:
        fx = f(x)
        fx_shifted = f(x + fx)
        theta = (fx_shifted - fx) / fx    # Compute theta(x_k)

        x_next = x - (fx / theta)    # Compute next x
        delta_x = x_next - x  # Change in x
        error_next = x_next - rho
        error = x - rho
        relative_error = error_next / error if error != 0 else np.nan  # Avoid division by zero
        iterations.append((k + 1, x_next, delta_x, relative_error))  # Log the iteration


        if abs(error_next) < epsilon:
            return x_next, iterations  # Convergence check

        x = x_next    # Update x
        k += 1

    raise ValueError("Steffensen's Method did not converge within the maximum number of iterations.")
```

```python
p1 = 0.00001
p2 = 0.0000001

def f3(x):
    f = x * (x-p1) * (x-p2)
    return f

def df3(x):
    df = 3 * (x**2) - 2 * (p1+p2) * x + (p1*p2)
    return df
```
